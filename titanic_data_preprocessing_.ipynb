```python
import os

folder_path = '/content/drive/My Drive/HDMxSp'
for filename in os.listdir(folder_path):
    print(filename)
```

```python
import pandas as pd

data_path = folder_path + '/Titanic_data.csv'
df = pd.read_csv(data_path)
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
df.fillna({'Age': df['Age'].median()}, inplace=True)
```

```python
df.info()
df.describe(include='all')
```

```python
df['Sex'] = df['Sex'].str.strip().str.lower()
# Only run this if Embarked still exists
if 'Embarked' in df.columns:
    df['Embarked'] = df['Embarked'].astype(str).str.strip().str.upper()
    df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)

```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
if 'Cabin' in df.columns:
    df.drop(columns=['Cabin'], inplace=True)
```

```python
print(df.columns)
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df['IsAlone'] = (df['FamilySize'] == 0).astype(int)
df.drop(columns=['Name', 'Ticket'], inplace=True)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Age', 'Fare', 'FamilySize']] = scaler.fit_transform(df[['Age', 'Fare', 'FamilySize']])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Titanic Data')
plt.show()
```

Let's visualize some relationships and distributions in the data.

```python
# Survival by Pclass
plt.figure(figsize=(6, 4))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Survival by Sex
plt.figure(figsize=(6, 4))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.ylabel('Survival Rate')
plt.xticks([0, 1], ['Male', 'Female'])
plt.show()
```

```python
# Survival by Embarked
plt.figure(figsize=(6, 4))
sns.barplot(x='Embarked_S', y='Survived', data=df) # Using one of the dummy variables
plt.title('Survival Rate by Embarked (Southampton vs Others)')
plt.ylabel('Survival Rate')
plt.xticks([0, 1], ['Not S', 'S'])
plt.show()
```

```python
# Distribution of Age
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], kde=True, bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Fare
plt.figure(figsize=(8, 5))
sns.histplot(df['Fare'], kde=True, bins=30)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```
